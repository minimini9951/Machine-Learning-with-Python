```python
# 라이브러리 로딩
import pandas as pd
import numpy as np

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
# 자동완성 속도 증가
%config Completer.use_jedi = False
```

```python
# csv 파일 로딩
structure = pd.read_csv(
    'C:/k_digital/source/data/노인여가+복지시설(구별)_20221113151448.csv')
human = pd.read_csv(
    'C:/k_digital/source/data/주민등록인구_20221113150419.csv')
```

```python
# structure 파일에서 필요한 부분만 추출
structure = structure.iloc[:, 0:2]
structure.head(3)
```

```python
# 두 데이터 프레임에 겹치는 부분이 있어서 열이름 통일
structure.insert(loc=0, column='동별', value=structure['자치구'])
structure = structure.drop('자치구', axis=1)
structure.head(3)
```

```python
# 데이터 프레임 확인
human.head(3)
```

```python
# 비율 열 추가
human['per'] = (human.iloc[:, 1] / human.iloc[:, 2]).apply(
    lambda x: np.round(x, 2))
```

```python
# 비율 크기순으로 정렬
human_val = human.iloc[1:, :].sort_values('per', ascending=False).reset_index(drop=True)
human_val.head(3)
```

```python
# 소계 행 추출
human_tot = human.iloc[0, :]
human_tot = pd.DataFrame(human_tot)
human_tot = human_tot.T
human_tot
```

```python
# 행 결합
human = pd.concat([human_tot, human_val]).reset_index(drop=True)
human
```

```python
# 열 결합
tot = pd.merge(human, structure)
```

```python
# 시설 비율 열 추가
tot['fac_per'] = ((tot.iloc[:, 4] / tot.iloc[0, 4]) * 100).apply(lambda x: np.round(x, 2))
tot
```

```python
# 노인 비율 으로 시설 비율을 나눔
tot['tot_per'] = ((tot.fac_per / tot.per)*100).apply(lambda x: np.round(x, 2))
tot
```

```python
# 소계의 첫번째의 값 삭제
tot.tot_per[0] = 0
tot
```

```python
# 최종 비율로 정렬
tot = tot.sort_values('tot_per', ascending=True).reset_index(drop=True)
tot
```

```python
# csv 파일로 추출
tot.to_csv("tot.csv", index=False)
```

```python
# 조건 추가(노인 비율이 평균 이상)
a = tot[tot.per >= tot.per[1:].mean()].reset_index(drop=True)
a
```

```python
# 조건 추가(최종 비율이 평균 이상)
b = a[tot.tot_per <= tot.tot_per[1:].mean()].reset_index(drop=True)
b
```

```python
# 조건에 모두 만족하는 csv 파일 추출
b.to_csv("complete.csv", index=False)
```
